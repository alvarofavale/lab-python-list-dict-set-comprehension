````python
#5.

products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
           inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
           return inventory

inventory = initialize_inventory(products)

def get_customer_orders():
    num_orders = int(input("Enter the number of products you want to order: "))

    customer_orders = set()
    
    for i in range(num_orders):
        product_name = input(f"Enter name of product to order ({i+1}/{num_orders}): ")
        if product_name in products:
            customer_orders.add(product_name)
        else:
            print("Sorry, choose from list:", products)
    
    return customer_orders

customer_orders=get_customer_orders()
customer_orders

def calculate_order_statistics (customer_orders, products):
    num_orders = len(customer_orders)
    unique_products = len(set(customer_orders))
    unique_ratio = unique_products / len(products)*100
    return num_orders, unique_ratio


def print_order_statistics(order_statistics):
    num_orders, unique_ratio = order_statistics
    print(f"Number of Orders: {num_orders}")
    print(f"Unique Product Ratio: {unique_ratio}")

order_statistics = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_statistics)

def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if product in inventory and inventory[product]>0:
            inventory[product]-=1
            
    updated_inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}


    return updated_inventory

print("Updated Inventory:")
for product, quantity in updated_inventory.items():
    print(f"{product}: {quantity}")

def total_price_customer_order (customer_orders):
    product_prices = {product:int(input(f"Enter price of product {product}: "))for product in customer_orders}
    total_prices = sum(product_prices.values())
    return total_prices

total_prices = total_price_customer_order (customer_orders)
print("Total price", total_prices)
````
